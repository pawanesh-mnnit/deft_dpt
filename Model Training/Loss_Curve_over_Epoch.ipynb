### Import Labraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Simulated loss values over epochs

```python
epochs = np.arange(1, 101)
main_loss = np.exp(-epochs / 50) + np.random.normal(0, 0.01, 100)
reg_loss = 0.1 * np.exp(-epochs / 60) + np.random.normal(0, 0.002, 100)
weighting_loss = 0.5 * np.exp(-epochs / 40) + np.random.normal(0, 0.003, 100)

# Plotting the loss curves
plt.figure(figsize=(15, 10))
plt.plot(epochs, main_loss, label="Main Loss ($\mathcal{L}_{main}$)", linewidth=2)
plt.plot(epochs, reg_loss, label="Regularization Loss ($\mathcal{L}_{reg}$)", linewidth=2)
plt.plot(epochs, weighting_loss, label="Weighting Loss ($\mathcal{L}_{weighting}$)", linewidth=2)

plt.xlabel("Epochs")
plt.ylabel("Loss Value")
plt.title("Loss Component Curves over Epochs")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Hyperparameter sensitivity table
sensitivity_data = {
    'Hyperparameter': ['η (eta)', 'ψ (psi)', 'κ (kappa)'],
    'Values Tried': ['0.05, 0.1, 0.2', '0.5, 1.0, 2.0', '0.1, 0.5, 1.0'],
    'Best Value': [0.1, 1.0, 0.5],
    'Impact Summary': [
        'Controls strength of regularization; too high restricts learning',
        'Penalizes deviation from identity transform; 1.0 gives best stability',
        'Regulates weighting sharpness; 0.5 balances focus and smoothness'
    ]
}

sensitivity_df = pd.DataFrame(sensitivity_data)
sensitivity_df
```
